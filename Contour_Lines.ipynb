```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.DataFrame(np.random.rand(10,10))
plt.figure(figsize=(8,6))
plt.contour(data.values)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Contour Lines with Colormap")
plt.colorbar(label="value")
plt.show()
```
